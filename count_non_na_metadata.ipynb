```python
metadata_fields = ['sex','mark','study_year','date','department']
```

```python
metadata_values = Document.objects.all().values(*metadata_fields)
```

```python
metadata_values
```

```python
import pandas as pd
```

```python
metadata_frame = pd.DataFrame(metadata_values)
```

```python
metadata_frame.count()
```

```python
metadata_frame.count().apply(lambda x: x/len(metadata_values))
```

```python
absol = metadata_frame.count().rename('Количество текстов')
rel = metadata_frame.count().apply(lambda x: x/len(metadata_values)).rename('Доля текстов')
```

```python
result = pd.DataFrame((absol,rel)).transpose()
```

```python
result.to_csv('non_na_prop.csv')
```
